# Matplotlib
Matplotlib is a plotting library for Python. It is used along with NumPy to provide an environment that is an effective open source alternative for MatLab. It can also be used with graphics toolkits like PyQt and wxPython.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y, c='g')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph in 2D")
plt.savefig('Test.png') # To save the diagram as png
```

```python
plt.plot(x,y, c='g')
```

## Creating Subplot

```python
plt.subplot(2,2,1) # fiest is row thwn column and then position
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g^-')
plt.subplot(2,2,3)
plt.plot(x,y,'r-*')
```

### Sine waves

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title('Sine wave form')
```

## Histogram

```python
f = np.array([1,56,85,45,4,6,9,10])
plt.hist(f, bins = 20)
```

The towers or bars of a histogram are called bins. The height of each bin shows how many values from that data fall into that range. The default value of the number of bins to be created in a histogram is 10. However, we can change the size of bins using the parameter bins in matplotlib

```python
f = np.array([1,56,85,45,4,6,9,10])
plt.hist(f, bins = 10)
```

## Box plot

```python
data = [np.random.normal(0,std,100) for std in range(1,10)]
plt.boxplot(data, vert = True,patch_artist= True)
    
```

The fill color of a matplotlib box plot is white by default, but you can override it setting patch_artist = True and passing a dict with facecolor to the boxprops argument, as in the following example. The default color for the median line is orange. However, you can change it passing a dict to medianpr.
vert : If True , the boxplot is drawn vertically (default). If False , it is drawn horizontally.
