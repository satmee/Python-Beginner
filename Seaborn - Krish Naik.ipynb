# Seaborn
Helps a lot with statistical data visualization and analysis. Best way to analyze Univariant and Bivariant Analysis.

## Ditribution Plot
1. distplot
2. jointplot - specifically for bivariant analysis
3. pairplot - more than 2 variables

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation using Heatmap
It only takes the columns with dtypes integer and float values, not object variables.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Jointplot

### Univariate Analysis

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

## Pairplot
To plot multiple pairwise bivariate distributions in a dataset, you can use the .pairplot() function. It allows us to plot pairwise relationships between variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'smoker')
```

```python
df['smoker'].value_counts()# to see count of a  speciific category
```

## Distplot
Helps to create Histogram

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde = False, bins = 10)
```

## Categorial Plot
1. boxplot

4. bar plot

## Countplot
Used mainly for counting

```python
sns.countplot(x = 'sex',data = df)
```

```python
sns.countplot(x = 'day',data = df)
```

```python
sns.countplot(x = 'smoker',data = df)
```

## Barplot
Needs both x and y parameter to be passed

```python
sns.barplot(x = 'sex', y =  'tip', data = df)
```

## Box plot
Represnted in box and whisker and presents information in five-number summary

```python
sns.boxplot(x='smoker', y='total_bill', data = df)
```

```python
sns.boxplot(x='day', y = 'total_bill', data = df, palette= 'rainbow')
```

```python
sns.boxplot(x='smoker', y='total_bill', data = df, hue ='sex')
```

## Violinplot
It helps us to visualize the distribution of data both in terms of Kernel Density Estimation and box plot

```python
sns.violinplot(x ='sex', y ='total_bill', palette = 'rainbow', data = df)
```

```python
sns.violinplot(x='day', y = 'total_bill', data = df, palette= 'rainbow')
```

```python
sns.violinplot(x='day', y = 'total_bill', data = df, palette= 'rainbow', hue='sex')
```

## Jointplot
Very helpful for univarian and bivariant graphs. It combines two type of chart, that is Scatter chart and histogram, where Scatter chart signifies the correlation between two variables and Histogram helps us visualize the distribution of data. So to understand both of them together, jointplot is used

```python
sns.jointplot(fd)
```

```python
sns.jointplot(x = 'sepal_length', y ='petal_length', data =fd,kind= 'hex')
```

## Violinplot
A violin plot plays a similar role as a box-and-whisker plot. It shows the distribution of data points after grouping by one (or more) variables. Unlike a box plot, each violin is drawn using a kernel density estimate of the underlying distribution.

```python
sns.violinplot(fd, palette='rainbow')
```
